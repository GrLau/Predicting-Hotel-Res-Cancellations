```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, f1_score
import xgboost as xgb
from imblearn.over_sampling import SMOTE
import numpy as np
from imblearn.under_sampling import RandomUnderSampler
import datetime
import shap
import matplotlib.pyplot as plt
```

### Load data

```python
df = pd.read_csv('Hotel Reservations.csv')

df['booking_status'] = df['booking_status'].replace({'Not_Canceled':0, 'Canceled':1})
```

```python
def get_day_of_week(row): 
    try:
        date = datetime.date(row['arrival_year'], row['arrival_month'], row['arrival_date']) 
    except:
        date = datetime.date(row['arrival_year'], row['arrival_month'], row['arrival_date']-1)

    return date.strftime('%A')
    


df['day_of_week'] = df.apply(get_day_of_week, axis=1)
```

```python
# perform one-hot encoding on the columns
one_hot = pd.get_dummies(df[['room_type_reserved', 'market_segment_type', 'type_of_meal_plan', 'day_of_week']])

# concatenate the one-hot encoding with the original dataframe
df = pd.concat([df, one_hot], axis=1)

# drop the original columns
df = df.drop(['room_type_reserved', 'market_segment_type', 'type_of_meal_plan', 'day_of_week'], axis=1)
```

### Post SHAP RF (TOP 4)

```python
#Get X and y
y = df['booking_status']
X = df[['lead_time', 'no_of_special_requests', 'avg_price_per_room', 'market_segment_type_Online']]


#Split X and y into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


# Number of trees in random forest
n_estimators = [500, 1000]

# Number of features to consider at every split
max_features = range(1,3,1)


# Minimum number of samples required to split a node
min_samples_split = range(50,150,10)

#Create Parameter grid
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'min_samples_split': min_samples_split}

# Create a base model
rf = RandomForestClassifier(class_weight = "balanced")

# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, verbose = 5, scoring=('f1_weighted'))

#Train grid_search
grid_search.fit(X_train, y_train)

#Identify best model
best_grid = grid_search.best_estimator_

#Identify best parameters
print(grid_search.best_params_)
```

```python
#Predictions
y_train_pred = best_grid.predict(X_train)
y_test_pred = best_grid.predict(X_test)

#Classification reports
train_report = classification_report(y_train, y_train_pred)
test_report = classification_report(y_test, y_test_pred)

print("Training Classification Report:\n", train_report)
print("Testing Classification Report:\n", test_report)

#Compute confusion matrix
train_cm = confusion_matrix(y_train, y_train_pred)
test_cm = confusion_matrix(y_test, y_test_pred)

#Plot Confusion Matrix
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Training Confusion Matrix
ConfusionMatrixDisplay(train_cm).plot(ax=ax[0])
ax[0].set_title('Training Confusion Matrix')

# Testing Confusion Matrix
ConfusionMatrixDisplay(test_cm).plot(ax=ax[1])
ax[1].set_title('Testing Confusion Matrix')

plt.show()
```
