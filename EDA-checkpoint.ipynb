### Prepare data

```python
df.isnull().sum()
```

```python
# perform one-hot encoding on the 'color' column
one_hot = pd.get_dummies(df[['room_type_reserved', 'market_segment_type', 'type_of_meal_plan']])

# concatenate the one-hot encoding with the original dataframe
df = pd.concat([df, one_hot], axis=1)

# drop the original 'color' column
df = df.drop(['room_type_reserved', 'market_segment_type', 'type_of_meal_plan'], axis=1)
```

```python
df
```

```python
df.info()
```

```python
def get_day_of_week(row): 
    try:
        date = datetime.date(row['arrival_year'], row['arrival_month'], row['arrival_date']) 
    except:
        date = datetime.date(row['arrival_year'], row['arrival_month'], row['arrival_date']-1)

    return date.strftime('%A')
    


df['day_of_week'] = df.apply(get_day_of_week, axis=1)
```

```python
sns.countplot(x = "day_of_week", hue = "booking_status", data = df)
plt.savefig('day_of_week.pdf')
```

```python
sns.boxplot(x = "day_of_week", y = "avg_price_per_room", hue = "booking_status", data = df)
plt.savefig('day_of_week X price.pdf')
```
